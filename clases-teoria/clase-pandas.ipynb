```python
import pandas as pd
import numpy as np
```

```python
# Lista de diccionario 
data  = {
    'nombre':['Ana','Maria','Carlos','Pedro', 'Laura'], 
    'edad': [25,32,28,45,29],
    'ciudad':['Madrid','Barcelona','Valencia','M치laga','Barcelona'],
    'salario':[4000,4500,55000,60000,70000]
}

df = pd.DataFrame(data)
print(df)
```

```python
print("3 primeras lineas ", df.head(3))

print("2 ultimas lineas ", df.tail(2))

print(df.info())

print(df.describe())
```

```python
nombres = df['nombre']
print(nombres)
sub_df = df[['nombre','edad']]
print(sub_df)
```

```python
#Filtrado
mayores_30 = df[df['edad'] > 30]
print(mayores_30)

#Personas Madrid
madrid = df[df['ciudad'] == 'Madrid']
print(madrid)

#usando query
barcelona = df.query("ciudad == 'Barcelona'")
print(barcelona)
```

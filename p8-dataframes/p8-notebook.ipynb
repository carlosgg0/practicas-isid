```python
# Ejercicio 6. Ejecuta el código proporcionado para crear el DataFrame df_ventas
# con 100 registros
np.random.seed(42)
datos_ventas = {
    'Fecha': pd.date_range('2024-01-01', periods=100, freq='D'),
    'Producto': np.random.choice(['Laptop', 'Mouse', 'Teclado', 'Monitor',
    'Tablet'], 100),
    'Cantidad': np.random.randint(1, 10, 100),
    'Precio_Unitario': np.random.choice([1200, 25, 80, 300, 450], 100),
    'Region': np.random.choice(['Norte', 'Sur', 'Este', 'Oeste'], 100),
    'Vendedor': np.random.choice(['Ana', 'Carlos', 'Maria', 'Pedro'], 100)
}
df_ventas = pd.DataFrame(datos_ventas)
df_ventas['Venta_Total'] = df_ventas['Cantidad'] * df_ventas['Precio_Unitario']

df_ventas
```

```python
# Todas las ventas de 'Laptop'
df_ventas[df_ventas["Producto"] == "Laptop"].head(5)
```

```python
# Ventas con monto total mayor a $1000
df_ventas[df_ventas["Venta_Total"] > 1000].head(5)
```

```python
# Ventas realizadas por 'Maria' en la región 'Norte'

df_ventas[(df_ventas["Vendedor"] == "Maria") & (df_ventas.Region == "Norte")]
```

```python
# Ejercicio 7. Manejo de valores nulos

# Crea una copia del DataFrame del ejercicio 6 e introduce valores nulos en 
# algunas celdas
df_ventas_nulls = df_ventas.copy()

np.random.seed(42)

for i in range(20):
    r = np.random.choice(df_ventas.shape[0])
    c = np.random.choice(df_ventas.shape[1])
    print(f"({r},{c})", df_ventas_nulls.iat[r, c])
    df_ventas_nulls.iat[r, c] = pd.NA

df_ventas_nulls
```

```python
# Identifica y cuenta los valores nulos por columna
df_ventas_nulls.isnull().sum()
```

```python
#Elimina las filas con valores nulos y compara el tamaño del DataFrame con
#el inicial

print(df_ventas_nulls.dropna().size)
print(df_ventas_nulls.size)
```

```python
categorical_cols = ["Region", "Vendedor", "Producto"]

for c in categorical_cols:
    mode = df_ventas_nulls[c].mode()[0]
    df_ventas_nulls[c] = df_ventas_nulls[c].fillna(mode)

df_ventas_nulls = df_ventas_nulls.ffill()

mean_cantidad = round(df_ventas_nulls["Cantidad"].mean())
df_ventas_nulls["Cantidad"] = df_ventas_nulls["Cantidad"].fillna(mean_cantidad)
```

```python
df_ventas_nulls["Precio_Unitario"] = df_ventas_nulls["Precio_Unitario"].fillna(
    df_ventas_nulls.groupby("Producto")["Precio_Unitario"].transform("mean")
)

df_ventas_nulls["Venta_Total"] = df_ventas_nulls["Cantidad"]*df_ventas_nulls["Precio_Unitario"]

df_ventas_nulls.isna().sum()
```

```python
# Ejercicio 8. Agrupaciones
# Calcula la venta total, promedio, máxima y número de ventas del total del
# DataFrame del ejercicio 6

print(df_ventas["Venta_Total"].mean())
print(df_ventas["Venta_Total"].max())
```

```python
# Agrupa por 'Producto' y calcula la suma de ventas por producto

df_ventas.groupby("Producto")["Cantidad"].sum()
```

```python
#Agrupa por 'Region' y calcula la venta total, promedio, máxima y número de
#ventas

df_ventas.groupby("Region").agg(
    Total_Dinero=("Venta_Total", "sum"),
    Promedio_Venta=('Venta_Total', 'mean'),
    Venta_Mas_Grande=('Venta_Total', 'max'),
    Numero_Ventas=('Venta_Total', 'count')
)
```

```python
# Identifica los top 3 vendedores por ventas totales

df_ventas.groupby("Vendedor").agg(
    Total_Ventas=("Venta_Total", "sum")
).sort_values(by="Total_Ventas", ascending=False).head(3)

df_ventas.groupby("Vendedor")["Venta_Total"].sum().sort_values().tail(3)
```

```python
# Ejercicio 9

# Crea en el DataFrame del ejercicio 6 una nueva columna 'Venta_Con_IVA'
# aplicando 21% de IVA

df_ventas["Venta_Con_Iva"] = df_ventas["Venta_Total"] * 1.21
df_ventas
```

```python
# Crea una columna 'Categoria_Precio' que clasifique como 'Alto' ( mayor que
# 500) o 'Bajo' (menor o igual que 500) según el precio


# df_ventas["Categoria_Precio"] = np.where(
#     df_ventas["Precio_Unitario"] > 500, 
#     "Alto",
#     "Bajo"
# ) Más eficiente

df_ventas["Categoria_Precio"] = df_ventas["Precio_Unitario"].apply(
    lambda x: "Alto" if x > 500 else "Bajo"
)
df_ventas
```

```python
# Usa apply() para crear una columna 'Tipo_Venta' que clasifique las ventas
# como 'Pequeña' (menos de 500), 'Mediana' (menos de 2000) o 'Grande' (en
# otro caso)

df_ventas["Tipo_Venta"] = df_ventas["Venta_Total"].apply(
    lambda x: "Pequeña" if x < 500 else "Mediana" if x < 2000 else "Grande" 
)

df_ventas
```

```python
# Ejercicio 10

info_producto = {
    'Producto': ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Tablet', 'Auriculares'],
    'Categoria': ['Electrónica', 'Accesorio', 'Accesorio', 'Electrónica',
    'Electrónica', 'Accesorio'],
    'Costo': [800, 15, 40, 200, 300, 20]
}
info_vendedores = {
    'Vendedor': ['Ana', 'Carlos', 'Maria', 'Pedro', 'Laura'],
    'Departamento': ['Ventas', 'Tecnología', 'Ventas', 'Marketing', 'Ventas'],
    'Salario_Base': [30000, 35000, 32000, 28000, 31000]
}

# Crea dos DataFrames adicionales: info_productos e info_vendedores

df_info_producto = pd.DataFrame(info_producto)
print(df_info_producto)

df_info_vendedores = pd.DataFrame(info_vendedores)
print(df_info_vendedores)
```

```python
# Realiza un INNER JOIN entre df_ventas y info_producto

df_ventas.merge(
    right=df_info_producto,
    how="inner",
    on="Producto"
)
```

```python
# Realiza un LEFT JOIN entre df_ventas y info_vendedores
df_nuevo = pd.merge(
    df_ventas, df_info_vendedores,
    how="left",
    on="Vendedor"
)

print(df_nuevo.query("Vendedor == 'Laura'")) # DataFrame vacío pq Laura no aparece en df_ventas
```
